**Bar** **chart**

```python
df = pd.DataFrame(data.head(20))

```

```python
# Grouping by 'Category'
grouped = df.groupby('Category')
```

```python
# Plot 1: Bar plot for Quantity by Product Name (Grouped by Category)
plt.figure(figsize=(10, 6))
for category, group in grouped:
    plt.bar(group['Product Name'], group['Quantity'], label=category)

    plt.xlabel('Product Name')
    plt.ylabel('Quantity')
    plt.title('Quantity by Product Name (Grouped by Category)')
    plt.xticks(rotation=45)
    plt.legend()
    plt.show()
```

**pie** **chart**

```python
# Generate a pie chart to visualize the distribution of values in the 'Discount' column
df['Discount'].value_counts().plot(kind='pie', autopct='%1.1f%%')

# Add a title to the pie chart indicating the variable being visualized
plt.title("Discount")

# Display the pie chart
plt.show()
```

**Line** **Graph**

```python
# Plot 3: Line plot for Original Price by Product Name
plt.figure(figsize=(10, 6))
for category, group in grouped:
    plt.plot(group['Product Name'], group['Original Price (Rs.)'], marker='o', label=category)

    plt.xlabel('Product Name')
    plt.ylabel('Original Price (Rs.)')
    plt.title('Original Price by Product Name (Grouped by Category)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

**Box** **plot** **graph**

```python
# Plot 4: Box plot for Discounted Price by Category
plt.figure(figsize=(8, 6))
df.boxplot(column='Discounted Price (Rs.)', by='Category')
plt.title('Discounted Price Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Discounted Price (Rs.)')
plt.suptitle('')
plt.show()
```

**scatter** **plot** **graph**

```python
# Plot 5: Scatter plot for Discount vs Discounted Price
plt.figure(figsize=(8, 6))
plt.scatter(df['Discount'], df['Discounted Price (Rs.)'], color='red')
plt.xlabel('Discount')
plt.ylabel('Discounted Price (Rs.)')
plt.title('Discount vs Discounted Price')
plt.grid(True)
plt.show()
```
